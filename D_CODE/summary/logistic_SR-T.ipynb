### Logistic Model - SR-T performance:

```python
# add the parent directory to the system path:
import sys
from pathlib import Path
parent_dir = Path('..').resolve() 
sys.path.append(str(parent_dir)) 

# import:
import numpy as np
import os
import matplotlib.pyplot as plt
import equations
from evaluation_trajectories import evaluation1D, evaluation1D_p, evaluation1D_2p 

# seed:
np.random.seed(666)
```

### Visualization

```python
# select the ODE:
ode_name = 'LogisticODE_k'
init_high_1 = [0.05, 0.3]
init_low_1 = [0.05, 0.3]

init_high_2 = [0.05, 0.5]
init_low_2 = [0.05, 0.5]

init_high_3 = [0.05, 0.7]
init_low_3 = [0.05, 0.7]

ode_param=None
x_id = 0
freq = 5
n_sample = 50
noise_ratio = 0.01
alg = 'diff'
seed = 0
n_seed = 5
N = 1

ode = equations.get_ode(ode_name, ode_param)
dt = 1 / freq
time_vector_b = np.arange(0, ode.T + dt, dt)
print(np.shape(time_vector_b))
```

```python
# evaluation: 
xt_true_1, est_trajectories_1, rmse_1 = evaluation1D_p(ode_name, ode_param, x_id, freq, n_sample, noise_ratio, alg, seed, n_seed, init_high_1, init_low_1, N)
print('RMSE on test-case 0: ', rmse_1)

xt_true_2, est_trajectories_2, rmse_2 = evaluation1D_p(ode_name, ode_param, x_id, freq, n_sample, noise_ratio, alg, seed, n_seed, init_high_2, init_low_2, N)
print('RMSE on test-case 1: ', rmse_2)

xt_true_3, est_trajectories_3, rmse_3 = evaluation1D_p(ode_name, ode_param, x_id, freq, n_sample, noise_ratio, alg, seed, n_seed, init_high_3, init_low_3, N)
print('RMSE on test-case 2: ', rmse_3)

# plot:
plt.figure(figsize=(5, 4))
for i in range(np.shape(est_trajectories_1)[0]):
    plt.plot(time_vector_b, est_trajectories_1[i][:, 0, 0], color='blue', linewidth=0.5, alpha=0.7)
plt.plot(time_vector_b, xt_true_1[:, 0, 0], color='red', linewidth=1.0, label='Correct Trajectory')
plt.scatter(time_vector_b[0], xt_true_1[:, 0, 0][0], color='green', label='Start')
plt.scatter(time_vector_b[-1], xt_true_1[:, 0, 0][-1], color='red', label='End')
plt.title('Estimated Trajectories - Param. $b$')
plt.legend()
plt.grid(True) 

plt.figure(figsize=(5, 4))
for i in range(np.shape(est_trajectories_2)[0]):
    plt.plot(time_vector_b, est_trajectories_2[i][:, 0, 0], color='blue', linewidth=0.5, alpha=0.7)
plt.plot(time_vector_b, xt_true_2[:, 0, 0], color='red', linewidth=1.0, label='Correct Trajectory')
plt.scatter(time_vector_b[0], xt_true_2[:, 0, 0][0], color='green', label='Start')
plt.scatter(time_vector_b[-1], xt_true_2[:, 0, 0][-1], color='red', label='End')
plt.title('Estimated Trajectories - Param. $b$')
plt.legend()
plt.grid(True)

plt.figure(figsize=(5, 4))
for i in range(np.shape(est_trajectories_3)[0]):
    plt.plot(time_vector_b, est_trajectories_3[i][:, 0, 0], color='blue', linewidth=0.5, alpha=0.7)
plt.plot(time_vector_b, xt_true_3[:, 0, 0], color='red', linewidth=1.0, label='Correct Trajectory')
plt.scatter(time_vector_b[0], xt_true_3[:, 0, 0][0], color='green', label='Start')
plt.scatter(time_vector_b[-1], xt_true_3[:, 0, 0][-1], color='red', label='End')
plt.title('Estimated Trajectories - Param. $b$')
plt.legend()
plt.grid(True)
```
